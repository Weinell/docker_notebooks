```python
# import all necessary libs
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("winemag-data-130k-v2.csv", index_col=0)

# Looking at first 5 rows of the dataset
df.head()
```

```python
print(f'{df.shape[0]:,} rows and {df.shape[1]} columns') # the :, is used to numerical formatting for thousands seperator
```

```python
# Groupby by country
country = df.groupby("country")

# Summary statistic of all countries
country.describe()
```

```python
print(f'{len(country)} countries')
print();
```

```python
# Select the top 5 highest average points
country.mean().sort_values(by="points",ascending=False).head()
```

```python
plt.figure(figsize=(15,10)) 
country.size().sort_values(ascending=False).plot.bar()
plt.xticks(rotation=50)
plt.xlabel("Country of Origin")
plt.ylabel("Number of Wines")
plt.show()
```

```python
text = "This wine is fruity and sweet, with bitter undertones and sweet notes of blueberry"
text = df.description[0]
print(text)
# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('danish'))
stop_words
```

```python
d_tokens = word_tokenize(d_txt)
d_tokens = [str(d) for d in d_tokens]
```

```python
import requests
res = requests.get('https://gist.githubusercontent.com/berteltorp/0cf8a0c7afea7f25ed754f24cfc2467b/raw/305d8e3930cc419e909d49d4b489c9773f75b2d6/stopord.txt')
stopwords = res.text
stopwords = stopwords.split('\n')
stopwords
```

```python
d_tokens = [t for t in d_tokens  if not t.lower() in stopwords]
d_tokens
```

```python
wordcloud = WordCloud().generate(' '.join(d_tokens))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
